```python
full_data = pd.read_csv('data.csv', encoding='latin-1')
print("summary")
full_data['content'].describe()
```

```python
#most common words from corpus which are words that are not significant in classification 
def get_common_word(corpus, n=None, stop_words=None):
    vectorizer = CountVectorizer(stop_words=stop_words).fit(corpus)
    bag_of_words = vectorizer.fit_transform(corpus)
    word_sum = bag_of_words.sum(axis=0) 
    common = [(word, word_sum[0, index]) for word, index in vectorizer.vocabulary_.items()]
    common =sorted(common, key = lambda x: x[1], reverse=True)
    return common[:n]

#common words in the corpus with stop words
common_word = get_common_word(full_data.content, 10)

#convert to panda dataframe 
common_df = pd.DataFrame(common_word)
#plot the result
plt.figure(figsize=(10,10))
plt.title('Common words included stop words')
sb.barplot(common_df[0], common_df[1], alpha=0.8)
plt.ylabel('Frequency', fontsize=12)
plt.xlabel('common words', fontsize=12)
plt.show()
```

```python
#common words in the corpus without stop word
common_no_stop = get_common_word(full_data.content, 10, 'english')

#convert to panda dataframe 
common_df = pd.DataFrame(common_no_stop)
#plot the result
plt.figure(figsize=(10,10))
plt.title('Common words excluded stop words')
sb.barplot(common_df[0], common_df[1], alpha=0.8)
plt.ylabel('Frequency', fontsize=12)
plt.xlabel('common words', fontsize=12)
plt.show()
```
