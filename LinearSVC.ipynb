```python
import csv
import numpy as np
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import seaborn as sb
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, f1_score
from keras.datasets import reuters
from sklearn.svm import LinearSVC, SVC
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
#get data from Keras 
max_features = 10000
(x_train, y_train), (x_test, y_test) = reuters.load_data(num_words=max_features)

#decode 
word_index = reuters.get_word_index()
revers_word_indx = dict([(value, key) for (key, value) in word_index.items()])

x_train = [
    ' '.join(revers_word_indx[i] for i in x_train[i] if i < len(revers_word_indx))for i in range(x_train.shape[0])
]

x_test = [
    ' '.join(revers_word_indx[i] for i in x_test[i] if i < len(revers_word_indx))for i in range(x_test.shape[0])
]
```

```python
#model 
model = Pipeline([
    ('tfidf', TfidfVectorizer(ngram_range=(1,2), stop_words='english', sublinear_tf=True, min_df=5, norm='l2', encoding='latin-1')),
    ('svc', LinearSVC())
])

history = model.fit(x_train, y_train)
print("done")
```

```python
#Model Accuracy
y_pred = model.predict(x_test)
print(classification_report(y_pred, y_test, labels=np.unique(y_pred)))
print((y_pred == y_test).mean())
```

```python
#confusioin matrix
conf_mat = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(15,15))
sb.heatmap(conf_mat, annot=True, fmt='d')
plt.ylabel('Actual Class')
plt.xlabel('Predicted Class')
plt.show()
```
